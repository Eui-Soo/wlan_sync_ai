```python
import numpy as np
import h5py
from keras.models import Model
from keras import models
from keras import layers
from keras import Input
from keras import optimizers
from keras.layers import Dropout
from keras.layers import BatchNormalization
import pydot
import graphviz
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
from keras.utils.vis_utils import plot_model
from keras.models import load_model
channel_model=['C','D','E','F']

import tensorflow as tf
from keras.backend.tensorflow_backend import set_session

config = tf.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.3
set_session(tf.Session(config=config))
```

```python
for model_idx in channel_model:
    model_name='wlan_timing_model_%s.h5' % model_idx
    model = load_model(model_name)
    SNR=range(-10,31)
    if model_idx=='B':
        First_peak_ACC_B=[]
        First_peak_ACC_B_v2=[]
        ACC_B=[]
    elif model_idx=='C':
        First_peak_ACC_C=[]
        First_peak_ACC_C_v2=[]
        ACC_C=[]
    elif model_idx=='D':
        First_peak_ACC_D=[]
        First_peak_ACC_D_v2=[]
        ACC_D=[]
    elif model_idx=='E':
        First_peak_ACC_E=[]
        First_peak_ACC_E_v2=[]
        ACC_E=[]
    elif model_idx=='F':
        First_peak_ACC_F=[]
        First_peak_ACC_F_v2=[]
        ACC_F=[]
    for SNR_loop in SNR:
        print(SNR_loop)
        ACC_tmp=0
        peak_tmp=0
        filepath='D:/논문연구/WLAN/wlan_test_set_Timing_model_{0}_SNR({1}).mat' .format(model_idx, SNR_loop)
        f=h5py.File(filepath,'r')
        test_data=np.array(f['Test_data'][:])
        test_time_label=np.array(f['Test_data_timing_label'][:])
        #a=model.evaluate(test_data, [test_time_label],batch_size=500)
        #if model_idx=='B':
        #    First_peak_ACC_B.append(a[1])
        #elif model_idx=='C':
        #    First_peak_ACC_C.append(a[1])
        #elif model_idx=='D':
        #    First_peak_ACC_D.append(a[1])
        #elif model_idx=='E':
        #    First_peak_ACC_E.append(a[1])
        #elif model_idx=='F':
        #    First_peak_ACC_F.append(a[1])        
        #print(a[1])
        for n in range(50000):
            temp=test_data[n]
            predict=np.argmax(model.predict(temp.reshape(1,11,864,1)))
            label=np.argmax(test_time_label[n])
            if model_idx=='B':
                if predict>=label-2 and predict <=label+2:
                    ACC_tmp=ACC_tmp+1
                    if predict==label:
                        peak_tmp=peak_tmp+1
            elif model_idx=='C':
                if predict>=label-3 and predict <=label+4:
                    ACC_tmp=ACC_tmp+1
                    if predict==label:
                        peak_tmp=peak_tmp+1
            elif model_idx=='D':
                if predict>=label-7 and predict <=label+8:
                    ACC_tmp=ACC_tmp+1
                    if predict==label:
                        peak_tmp=peak_tmp+1
                
            elif model_idx=='E':
                if predict>=label-14 and predict <=label+14:
                    ACC_tmp=ACC_tmp+1
                    if predict==label:
                        peak_tmp=peak_tmp+1
                
            elif model_idx=='F':
                if predict>=label-15 and predict <=label+15:
                    ACC_tmp=ACC_tmp+1
                    if predict==label:
                        peak_tmp=peak_tmp+1
                
            
        if model_idx=='B':
            ACC_B.append(ACC_tmp/50000)
            First_peak_ACC_B.append(peak_tmp/ACC_tmp)
            First_peak_ACC_B_v2.append(peak_tmp/50000)
        elif model_idx=='C':
            ACC_C.append(ACC_tmp/50000)
            First_peak_ACC_C.append(peak_tmp/ACC_tmp)
            First_peak_ACC_C_v2.append(peak_tmp/50000)
        elif model_idx=='D':
            ACC_D.append(ACC_tmp/50000)
            First_peak_ACC_D.append(peak_tmp/ACC_tmp)
            First_peak_ACC_D_v2.append(peak_tmp/50000)
        elif model_idx=='E':
            ACC_E.append(ACC_tmp/50000)
            First_peak_ACC_E.append(peak_tmp/ACC_tmp)
            First_peak_ACC_E_v2.append(peak_tmp/50000)
        elif model_idx=='F':
            ACC_F.append(ACC_tmp/50000)   
            First_peak_ACC_F.append(peak_tmp/ACC_tmp)
            First_peak_ACC_F_v2.append(peak_tmp/50000)
np.save('model_B_timing_FER.npy',ACC_B)
np.save('model_B_timing_ACC_peak.npy',First_peak_ACC_B)
np.save('model_B_timing_ACC_peak_v2.npy',First_peak_ACC_B_v2)
np.save('model_C_timing_FER.npy',ACC_C)
np.save('model_C_timing_ACC_peak.npy',First_peak_ACC_C)
np.save('model_C_timing_ACC_peak_v2.npy',First_peak_ACC_C_v2)
np.save('model_D_timing_FER.npy',ACC_D)
np.save('model_D_timing_ACC_peak.npy',First_peak_ACC_D)
np.save('model_D_timing_ACC_peak_v2.npy',First_peak_ACC_D_v2)
np.save('model_E_timing_FER.npy',ACC_E)
np.save('model_E_timing_ACC_peak.npy',First_peak_ACC_E)
np.save('model_E_timing_ACC_peak_v2.npy',First_peak_ACC_E_v2)
np.save('model_F_timing_FER.npy',ACC_F)
np.save('model_F_timing_ACC_peak.npy',First_peak_ACC_F)
np.save('model_F_timing_ACC_peak_v2.npy',First_peak_ACC_F_v2)
```
